# Creating a custom Huber loss func and training a NN model using it.

## Importing libraries

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt
from ipynb.fs.full.Data_processing import data_pipeline # For getting the processed data
from sklearn.datasets import fetch_california_housing
sns.set(style = 'whitegrid')
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import selu
from tensorflow.keras.initializers import lecun_normal
from tensorflow.keras.optimizers import Nadam
```

## Loading the data

```python
housing = fetch_california_housing()
x_train, x_train_scaled, x_valid, x_valid_scaled, x_test, x_test_scaled, y_train, y_valid, y_test = data_pipeline(housing)
# A predefined func in another notebook is used to get the data, split and scale it.
```

## Huber loss func

```python
def huber_loss(y_true, y_pred):
    error = y_true - y_pred
    small_error = tf.abs(error) < 1
    squared_error = tf.square(error) / 2
    linear_error = tf.abs(error) - 0.5
    return tf.where(small_error, squared_error, linear_error)
```

```python
# Plotting Huber func
z = np.linspace(-4, 4, 200)
plt.figure(figsize = (10, 10))
sns.lineplot(z, huber_loss(0, z), label = 'Huber_fn')
sns.lineplot(z, tf.square(z) / 2, label = r"$\frac{1}{2}z^2$")
plt.axis([-4.5, 4.5, -0.5, 4])
```

## Model

```python
# For clearing previous model history and setting the seed gen values
def pre_model():
    keras.backend.clear_session()
    tf.random.set_seed(42)
    np.random.seed(42)
```

```python
pre_model()
```

```python
# For compiling and training the model
def compile_train(model, train_data, validation_data, loss, optimizer, metrics, epochs):
    model.compile(loss = loss, optimizer = optimizer, metrics = metrics)
    history = model.fit(train_data[0], train_data[1], epochs = epochs, validation_data = validation_data)
    return history
```

```python
input_shape = x_train.shape[1:]
```

```python
model = Sequential()
model.add(Dense(30, input_shape = input_shape, activation = selu, kernel_initializer = lecun_normal()))
model.add(Dense(1))
```

```python
model.summary()
```

```python
history = compile_train(model, (x_train_scaled, y_train), (x_valid_scaled, y_valid), huber_loss, Nadam(), 'accuracy', 5)
```

## Saving and loading models with custom objects

```python
model.save('Huber_loss_model')
```

```python
model = keras.models.load_model('Huber_loss_model', custom_objects = {'huber_loss' : huber_loss})
```

```python
model.fit(x_train_scaled, y_train, epochs = epochs, validation_data = (x_valid, y_valid))
```

- The above huber func has a permanent threshold setting of 0. The threshold value should be tunable.

```python
# Huber func with modifiable threshold value
def create_huber(threshold = 1.):
    def huber_loss(y_true, y_pred):
        error = y_true - y_pred
        small_error = tf.abs(error) < threshold
        squared_error = tf.square(error) / 2
        linear_error = threshold * tf.abs(error) - tf.square(threshold) / 2
        return tf.where(small_error, squared_error, linear_error)
    return huber_loss
```

```python
history = compile_train(model, (x_train_scaled, y_train), (x_valid_scaled, y_valid), create_huber(2.), Nadam(), 'accuracy', 5)
```

```python
model.save('Modifiable_Huber_loss_model')
```

```python
model = keras.models.load_model('Modifiable_Huber_loss_model', custom_objects = {'huber_fn' : create_huber(2.)})
```

```python
model.fit(x_train_scaled, y_train, epochs = epochs, validation_data = (x_valid, y_valid))
```

## Subclassing the loss class in keras

- Since loading the model along with the threshold is an issue, we can subclass the loss module in keras to create a custom loss class.

```python
class huber_class(keras.losses.Loss):
    def __init__(self, threshold = 1., **kwargs):
        self.threshold = threshold
        super().__init__(**kwargs)
    def call(self, y_true, y_pred): # Huber loss func
        error = y_true - y_pred
        small_error = tf.abs(error) < self.threshold
        squared_error = tf.square(error) / 2
        linear_error = self.threshold * tf.abs(error) - tf.square(self.threshold) / 2
        return tf.where(small_error, squared_error, linear_error)
    def get_config(self): # Saving threshold along with model
        base_config = super().get_config()
        return {**base_config, 'threshold' : self.threshold}
```

```python
model.summary()
```

```python
history = compile_train(model, (x_train_scaled, y_train), (x_valid_scaled, y_valid), huber_class(2.), Nadam(), 'mae', 5)
```

```python
model.save('Subclassed_Huber_model')
```

```python
model = keras.models.load_model('Subclassed_Huber_model', custom_objects = {'huber_class' : huber_class})
```

```python
model.fit(x_train_scaled, y_train, epochs = 5, validation_data = (x_valid_scaled, y_valid))
```
