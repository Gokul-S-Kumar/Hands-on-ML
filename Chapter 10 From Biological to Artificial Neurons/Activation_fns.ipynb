# Plotting activation functions and its derivatives

## Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'whitegrid')
```

### Sigmoid function

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

### ReLU function

```python
def relu(z):
    return np.maximum(0, z)
```

### Tanh function

```python
def tanh(z):
    return np.tanh(z)
```

### Step function

```python
def step(z):
    return np.sign(z)
```

### Derivative function

```python
def derivative(f, z, eps = 0.000001):
    return (f(z + eps) - f(z - eps)) / (2 * eps)
```

```python
z = np.linspace(-5, 5, 200)
```

## Plotting the functions

```python
fig = plt.figure(figsize = (10,10))
ax = plt.plot(z, sigmoid(z), label = 'Sigmoid', linewidth = 3)
ax = plt.plot(z, relu(z), label = 'ReLU', linewidth = 3)
ax = plt.plot(z, tanh(z), label = 'Tanh', linewidth = 3)
ax = plt.plot(z, step(z), label = 'Step', linewidth = 3)
plt.axis([-5, 5, -1.2, 1.2])
plt.legend()
```

## Plotting its derivatives

```python
fig = plt.figure(figsize = (10,10))
ax = plt.plot(z, derivative(sigmoid, z), label = 'Sigmoid', linewidth = 3)
ax = plt.plot(z, derivative(relu, z), label = 'ReLU', linewidth = 3)
ax = plt.plot(z, derivative(tanh, z), label = 'Tanh', linewidth = 3)
ax = plt.plot(z, derivative(step, z), label = 'Step', linewidth = 3)
plt.axis([-5, 5, -1.2, 1.2])
plt.legend()
```

- We can see that the derivative of step function is zero. So calculating gradients using step function as activation function is not possible.
